```python
#FACTORIAL
def fac(n):
    # Initialize x
    x = 1
    
    # Deal with edge cases, otherwise calculate factorial
    if n < 0:
        print("Undef.")
    elif n == 0:
        print("0! = 1")
    else:
        for i in range(1,n+1):
                x = x * i
        print(x)
                
fac(3)
```

```python
#ODD_PRODUCT
def odd_product(n):
    # Compares remainders of products to zero
    for i in range ( 0, len(n)):
        for j in range ( i+1 , len(n)):
            if( (n[i]*n[j])%2 !=0 and n[i]!=n[j] ):
                return True

    return False

x = [2,4,5]
y = [1,3,4]

print(odd_product(x))
print(odd_product(y))
```

```python
#REVERSE_INTEGER
def reverse_integer(n):
    # Convert int to string, reverse it, and multiply by -1 if n<0
    x = str(abs(n))
    x = x[::-1]
    x = int(x)

    if n < 0:
        print(-1*x)
    else:
        print(x)
    
reverse_integer(-123)
```

```python
#REMOVE_COMMAS
def removecommas(sen):
    # Replace commas with nothing
    sen = sen.replace(',', '')
    print(sen)
    
removecommas("Sit down, please")
```

```python
#VALID_PARENTHESES

def check(string): 
    # Check string for bracket, then replace matching parentheses
    brackets = ['()', '{}', '[]'] 
    while any(x in string for x in brackets): 
        for types in brackets: 
            string = string.replace(types, '') 
        return not string

check('[{}][()]')
```

```python
#MERGE_SORTED_LIST

def merge_sorted(a,b):
    # Read lengths of objects
    x = len(a)
    y = len(b)
    
    # Initialize objects
    merged = []
    i = 0
    j = 0
    
    # If created objects are shorter than originals, append to merged list
    while i < x and j < y: 
        if a[i] < b[j]: 
            merged.append(a[i]) 
            i += 1

        else: 
            merged.append(b[j]) 
            j += 1
  
    merged = merged + a[i:] + b[j:]
    print(merged)
    
merge_sorted([1,3,4],[1,2,6,8])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Writing a function to plot any equation
def plot(equation, domain):  
    x = np.array(domain)  
    y = equation(x)
    plt.plot(x, y)  
    plt.show()  
    
# Using function to plot different equations
plot(lambda x: 30+0.5*x, range(0,10))
plot(lambda x: (x-25)**2-20, range(0,50))
plot(lambda x: -np.log(x), range(1,10))
plot(lambda x: -np.log(1-x), range(-10,0))
plot(lambda x: 1/(1+np.exp(-x)), range(0,10))
```
